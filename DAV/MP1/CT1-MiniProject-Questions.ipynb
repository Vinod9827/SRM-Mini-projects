```python
#import libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

### 1. Read the dataset 'pokemon.csv'.  [1 mark]

```python
df = pd.read_csv("Pokemon.csv")
df
```

### 2.List the columns in the dataframe.    [2 marks]

```python
df.columns
```

### 3.Drop the column '#'.     [1 mark]

```python
df = df.drop(columns="#")
df
```

### 4.Check if the dataset has null values.    [2 marks]

```python
# to check with respect to each data
df.isnull()

# to check the number of null values present in each column
df.isnull().sum()

# to check the number of null values present in dataset
total_number = df.isnull().sum().sum()

# therefore

if (total_number != 0):
    print( "The dataset have",total_number,"null values")
else:
    print("The dataset doesn't have null values")
```

### 5.Drop the records with null values. [2 marks]

```python
df = df.dropna()
df
```

### 6. Find the most frequent type of pokemon in Type 1 and Type 2.  [2 marks]

```python
df[['Type 1','Type 2']].mode()
```

### 7. Find the mean defense value and attack value of the pokemons.  [2 marks]

```python
df[['Attack','Defense']].mean()
```

### 8. What is the maximum 'total' of the pokemons? List the pokemons with top 5 total. [2 marks]

```python
df.sort_values(by=['Total'],ascending=False, inplace=True)
df["Name"].head(5)
```

### 9.How skewed are the features 'Sp. Def' and 'Sp. Atk'? Write down your inference of the same.  [4 marks]

```python
fig = plt.subplots(figsize=(10,5))
plt.subplot(1, 2, 1)

sns.distplot(df['Sp. Atk'])
plt.subplot(1, 2, 2)
sns.distplot(df['Sp. Def'])
plt.show()
print ("Shew Values for \nSp.Atck :",df['Sp. Atk'].skew(),"\nSp.Def :",df['Sp. Def'].skew())

print("Inference : As part of plots we can figure out that the 'Sp. Atk' and 'Sp. Def' are been postive skewed. Therefore, mean > median > mode.")
```

### 10.Plot a histogram for the feature 'Speed'. What are your inferences from the same? [2 marks]

```python
plt.hist(x = df['Speed'])
plt.xlabel('Speed')
plt.ylabel('No of Pokemons')
plt.show()
print("Inference : ")
print("Least number of Pokemons have the top speed between 140-160")
print("Max Number of Pokemons have the speed comes under the top speed between 75-80")
```
